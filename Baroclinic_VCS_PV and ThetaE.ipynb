```python
import matplotlib.pyplot as plt
import metpy.calc as mpcalc
from metpy.interpolate import cross_section
import numpy as np
import xarray as xr
```

```python
# Open each dataset from 00 to 09Z of ERA5 Re; change name accordingly to yours
ds1 = xr.open_dataset('EU_VCS_00.nc')
ds1 = ds1.metpy.parse_cf().squeeze()
ds2 = xr.open_dataset('EU_VCS_03.nc')
ds2 = ds2.metpy.parse_cf().squeeze()
ds3 = xr.open_dataset('EU_VCS_06.nc')
ds3 = ds3.metpy.parse_cf().squeeze()
ds4 = xr.open_dataset('EU_VCS_09.nc')
ds4 = ds4.metpy.parse_cf().squeeze()
```

```python
start = (10, 115)
end = (20, 125)

#------------------------EU_VCS_00.nc4------------------------#

cross1 = cross_section(ds1, start, end).set_coords(('latitude', 'longitude'))
lon1 = cross1['longitude'] * units.degrees_east
lat1 = cross1['latitude'] * units.degrees_north
pres1 = cross1['level'] * units('hPa')
temp1 = cross1['t'] * units('K')

#------------------------EU_VCS_03.nc4------------------------#

cross2 = cross_section(ds2, start, end).set_coords(('latitude', 'longitude'))
lon2 = cross2['longitude'] * units.degrees_east
lat2 = cross2['latitude'] * units.degrees_north
pres2 = cross2['level'] * units('hPa')
temp2 = cross2['t'] * units('K')

#------------------------EU_VCS_06.nc4------------------------#

cross3 = cross_section(ds3, start, end).set_coords(('latitude', 'longitude'))
lon3 = cross3['longitude'] * units.degrees_east
lat3 = cross3['latitude'] * units.degrees_north
pres3 = cross3['level'] * units('hPa')
temp3 = cross3['t'] * units('K')

#------------------------EU_VCS_09.nc4------------------------#

cross4 = cross_section(ds4, start, end).set_coords(('latitude', 'longitude'))
lon4 = cross4['longitude'] * units.degrees_east
lat4 = cross4['latitude'] * units.degrees_north
pres4 = cross4['level'] * units('hPa')
temp4 = cross4['t'] * units('K')
```

```python
# Compute for Potential Temperature to each dataset.
pv1 = cross1['pv']
pot_temp1 = mpcalc.potential_temperature(pres1, temp1)

pv2 = cross2['pv']
pot_temp2 = mpcalc.potential_temperature(pres2, temp2)

pv3 = cross3['pv']
pot_temp3 = mpcalc.potential_temperature(pres3, temp3)

pv4 = cross4['pv']
pot_temp4 = mpcalc.potential_temperature(pres4, temp4)
```

```python
pv1_range = np.arange(-0.000005, 0.000005, 0.0000005)
pv2_range = np.arange(-0.000002, 0.0000072, 0.0000005)
pv3_range = np.arange(-0.000002, 0.0000072, 0.0000005)
pv4_range = np.arange(-0.000002, 0.0000055, 0.0000005)
```
